```python
# used b/w the numerical and the string -type value, for ex= total_bill and  day

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data=sns.load_dataset("tips")
print(data)
# sns.stripplot(data=data,x="sex",y="tip",hue="sex")
sns.stripplot(data=data,x="sex",y="tip",hue="time",dodge=True,jitter=0.9)
# hue= it is used to represent the symbol
# jitter=it is used to separate the two points, i.e- more jitter value, more will be the space between the  two points
# dodge= it is used to separte between the two things
plt.show()             
```

```python
# by scatter plot
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

data=sns.load_dataset("tips")
sns.scatterplot(data=data,x="day",y="total_bill")
plt.show()             
```

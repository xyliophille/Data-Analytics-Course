```python
import matplotlib.pyplot as plt

brands=["oneplus","apple","samsung","nokia","redmi"]
x=[22,35,30,3,10]
c=["red","silver","blue","pink","orange"]
ex=[0,0,0,0,0.1]
plt.pie(x,labels=brands,colors=c,explode=ex,shadow=True,autopct="%.2f",startangle=90) 
# plt.pie(x,labels=brands,colors=c,explode=ex) 
# explode means the highlighted region comes out from the graoh
#startangle is in degrees and it tells from where the series start 
# autopct tells % of each region, .f= without decimal and .2f =with decimal
plt.show()
```

```python
import pandas as pd 
import matplotlib.pyplot as plt

data=pd.read_excel("")
df=pd.DataFrame(data) #yha tk data list aayega output me

grouped_by=df.group_by("payment mode")["amount"].sum()
print(grouped_by)
plt.pie(grouped_by.values,labels=grouped_by.index,autopct="%.2f")
plt.show()
print(df)
```

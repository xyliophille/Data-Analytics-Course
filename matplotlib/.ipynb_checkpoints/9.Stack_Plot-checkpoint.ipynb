```python
import matplotlib.pyplot as plt
days=[1,2,3,4,5,6,7]

NOP1=[5,10,15,20,56,35,33]
NOP2=[10,20,30,40,50,90,89]
NOP3=[8,30,50,60,70,90,100]

plt.stackplot(days,NOP1,NOP2,NOP3,baseline="zero",colors=["pink","blue","red"],labels=["week1","week2","week3"])
plt.legend()
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt

data =pd.read_excel("")
df=pd.DataFrame(data)
# print(df)
grouped=df.groupby("category")["calories","protein","fats"].agg("mean")
plt.stackplot(df["category"],unique(),grouped["calories"],grouped["protein"],grouped["fats"])
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[45,34,56,23,45]
# creating subplot (no.of rows, no.of columns, no. of figure)
plt.subplot(1,2,1) 
plt.title("age")
plt.plot(x,y)
# plt.bar(x,y) ---> for bar ,graph 


x=[5,6,7,8,9]
y=[89,78,57,88,76]
plt.subplot(1,2,2)
plt.title("weight")
plt.plot(x,y)
# plt.suptitle("employee data")
plt.show()
```

```python
# practice

import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[45,34,56,23,45]
# creating subplot (no.of rows, no.of columns, no. of figure/charts)
plt.subplot(2,2,1)  
plt.plot(x,y)


x=[5,6,7,8,9]
y=[89,78,57,88,76]
plt.subplot(2,2,2)
plt.scatter(x,y)

x=[4,8,9,1,5]
y=[67,89,45,87,80]
plt.subplot(2,2,3)
plt.bar(x,y)

x=[4,8,9,1,5]
y=[67,89,45,87,80]
plt.subplot(2,2,4)
plt.step(x,y)
plt.show()
```

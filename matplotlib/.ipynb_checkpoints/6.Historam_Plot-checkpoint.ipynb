```python
# it is used to know the frequency of data

import matplotlib.pyplot as plt

x= [30,40,50,60,40,90,50,56,56,35,42,89,54,47,32,56,11,70,65]
plt.hist(x,bins =15 ,edgecolor= "black",color="pink")
plt.show()



# note:-
# What are bins?
# Bins are the intervals that divide the entire range of values into a series of consecutive, non-overlapping intervals. In your histogram:

# Your data appears to range from about 0 to 90 on the x-axis

# With bins=15, the histogram will group your data into 15 equally spaced vertical bars (rectangles) between the minimum and maximum values
```

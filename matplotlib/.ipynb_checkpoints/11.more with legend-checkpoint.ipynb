```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[45,67,33,67,12]
y1=[41,60,13,66,13]
y2=[11,30,55,60,19]
y3=[11,50,45,90,19]

plt.figure(figsize=[5,3])
plt.plot(x, y,label="male")
plt.plot(x,y1,label="female")
plt.plot(x,y2,label="kids")
plt.plot(x,y3,label="blanks")

# plt.legend()

plt.legend(["a1","a2","a3","a4"],ncols=2)
# plt.legend(loc=0) #loc is used to change the location of legend
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[45,67,33,67,12]
y1=[41,60,13,66,13]
y2=[11,30,55,60,19]
y3=[11,50,45,90,19]

plt.figure(figsize=[5,3])
plt.plot(x, y,label="male")
plt.plot(x,y1,label="female")
plt.plot(x,y2,label="kids")
plt.plot(x,y3,label="blanks")

# plt.legend()

# plt.legend(bbox_to_anchor=(0.8,1.2),ncols=2) # to change the location of the legend
plt.legend(bbox_to_anchor=(0.8,1.2),ncols=2,labelspacing=2) # to give the space between the legend
# plt.legend(loc=0) #loc is used to change the location of legend
plt.show()
```
